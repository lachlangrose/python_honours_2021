# Programming basics
## Variables
A variable is something that can be used to store data. In most programming languages variables are either mutable or immutable, this means the value in the variable can be changed or can't be changed. A nice analogy for a variable is a backpack, you might only be able to keep one book in your backpack so at the start of the day you put your math text book in it. To go to your English class you need to remove the math book and put the English text book in it. The key point to remember is that the contents of the backpack changes but we can always refer to the backpack. 

## Booleans
True or False statements (Booleans) are used a lot inside computer code. Booleans can either be expressed as variables or as part of a statement.  Booleans can be combined together where they can be combined using an OR or AND statement. Or means that if either boolean is True then the result is True, AND statement mean for the result to be True both booleans must be true.

## Control flow
Control flow is how we can tell the computer to do different things depending on a condition think about is as a traffic controller for your code.  
### Conditionals
A conditional statement will evaluate an expression into a boolean and determine how to proceed depending on if the result is True or False

```
variable a
variable b

a = random_number(0,10)
b = random_number(0,10)

if a + b == 10:
    print('a+b == 10')

else:
    print('a + b != 10')
```    
#### Selection
Selection is where there are multiple true of false statements to determine which part of code is executed.

```
variable a
variable b

a = random_number(0,10)
b = random_number(0,10)

#selection
if a + b == 10:
    print('a+b == 10')
else if a + b == 0:
    print('a + b is 0')
else if a + b == 10:
    print('this never happens')
else: # catches everything else
    print('a + b != 10')
```    
   
#### Nested
We can also have mutiple conditional statements that are nested inside other statements.

```
variable a
variable b

a = random_number(0,10)
b = random_number(0,10)

#selection
if a + b == 10:
    print('a+b == 10')
    else random_number(0,10) == 0:
        print('a + b is 10 and random number is 0')
```
